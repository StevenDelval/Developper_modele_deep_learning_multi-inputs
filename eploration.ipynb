```python
import pandas as pd
import re
```

```python
df = pd.read_csv("Data/flipkart_com-ecommerce_sample_1050.csv")
```

```python
df.head()
```

```python
print(df[df["image"]=="1120bc768623572513df956172ffefeb.jpg"]["description"].values)
```

```python
df["product_category_tree"].value_counts()
```

```python
df["product_category_tree_list"] = df["product_category_tree"].str.replace('"',"")
df["product_category_tree_list"] = df["product_category_tree_list"].str.replace('[',"")
df["product_category_tree_list"] = df["product_category_tree_list"].str.replace(']',"")
```

```python
df["product_category_tree_list"].unique()
```

```python
df["product_category_tree_target"] = df["product_category_tree_list"].str.split(">>")
```

```python
df["product_category_tree_target"] = df["product_category_tree_target"].apply(lambda x:x[0].rstrip().lstrip() + " > " +x[1].rstrip().lstrip())
```

```python
df["product_category_tree_target"].value_counts()
```

```python
df.drop(["product_category_tree_list","product_category_tree"],axis=1,inplace=True)
```

```python
df  = df[["image","product_name","description","product_category_tree_target"]]
```

```python
df["name_and_description"] = df.apply(lambda x : "Product name : " + x["product_name"].rstrip().lstrip() + "\nDescription : " + re.sub(r'\s+', ' ', x["description"]).rstrip().lstrip().strip().replace("\n"," ").replace("\t","").replace("\r",""),axis=1)
```

```python
df  = df[["image","name_and_description","product_category_tree_target"]]
```

```python
df.to_csv("./Data/preparation.csv",index=False)
```
